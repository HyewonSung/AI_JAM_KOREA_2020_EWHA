```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/user/Desktop/finished_petition_csv.csv")
```

```python
def find_day(count):
    return count[:-3]
```

```python
df
```

```python
# 데이터 시작일 년-월 형식으로 바꾸기

df["시작일"] = df["시작일"].map(find_day)
df
```

```python
# 중복된 월 제거

months = list(set(df["시작일"]))
months
```

```python

months_titles = []

for month in months:
    m_list = list(df[df["시작일"]==month]["제목"])
    m_list = ' '.join(m_list)
    months_titles.append(m_list)
```

```python
months_titles
```

```python
len(months_titles)
```

```python
# 월별로 원하는 키워드가 몇 번 나왔는지 보여주는 함수

def fined_month_keyword(word):
    count_list = []
    for months_title in months_titles:
        count_list.append(months_title.count(word))
    return count_list
```

```python
te1 = fined_month_keyword("교육")
te1
```

```python
te2 = fined_month_keyword("미세먼지")
te2
```

```python
ms = []
for m in months:
    ms.append(m[3:4]+m[5:])
ms
```
